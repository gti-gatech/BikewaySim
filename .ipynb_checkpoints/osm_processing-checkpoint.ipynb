## Choose Study Area

## Download OSM from OSMnx
### More info about OSMnx: https://geoffboeing.com/2016/11/osmnx-python-street-networks/
### OSMnx documentation: https://osmnx.readthedocs.io/en/stable/index.html

```python
osmnx_nodes, osmnx_links = download_osmnx(gdf)
```

```python
osmnx_links.head()
```

## Download OSM as JSON from Overpass API 

```python
overpass_links = overpass_download(gdf)
overpass_links.head()
```

```python
#get rid of duplicate columns names
overpass_links = overpass_links.iloc[:,overpass_links.columns.str.lower().duplicated(keep='first') == False]
```

### Export Overpass API data as a Pickle
#### Note: This takes a while

```python
pickle.dump(overpass_links, open("osm/osm_links_attr.p","wb"))
```

## Add highway column to osmnx links
#### Highway column needed for the filtering process

```python
osmnx_links = pd.merge(osmnx_links, overpass_links[['id','highway']], left_on=('osmid'), right_on=('id'), how = 'inner')
display(osmnx_links)
```

## Project to desired CRS

```python
osmnx_links.to_crs(epsg='2240', inplace = True)
osmnx_nodes.to_crs(epsg='2240', inplace = True)
```

## Export links and nodes as GeoJSON

```python
osmnx_links.to_file(f'osm/osm_links.geojson', driver = 'GeoJSON')
osmnx_nodes.to_file(f'osm/osm_nodes.geojson', driver = 'GeoJSON')
```
